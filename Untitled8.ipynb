```python
import os

class WaterTracker:
    def __init__(self):
        self.goal = 0
        self.logs = []
        self.load_logs()

    def set_goal(self, ml):
        self.goal = ml
        print(f"Daily water goal set to {ml} ml.\n")

    def add_entry(self, ml):
        self.logs.append(ml)
        self.save_logs()
        print(f"Logged {ml} ml of water.\n")

    def get_total(self):
        return sum(self.logs)

    def get_remaining(self):
        return max(0, self.goal - self.get_total())

    def get_percentage(self):
        if self.goal == 0:
            return 0
        return (self.get_total() / self.goal) * 100

    def save_logs(self):
        with open("water_log.txt", "w") as file:
            for log in self.logs:
                file.write(str(log) + "\n")

    def load_logs(self):
        if os.path.exists("water_log.txt"):
            with open("water_log.txt", "r") as file:
                self.logs = [int(line.strip()) for line in file.readlines()]

def main():
    tracker = WaterTracker()

    while True:
        print("=== Water Intake Tracker ===")
        print("1. Set Daily Goal")
        print("2. Log Water Intake")
        print("3. View Report")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            ml = int(input("Enter daily goal (ml): "))
            tracker.set_goal(ml)

        elif choice == "2":
            ml = int(input("Enter water intake (ml): "))
            tracker.add_entry(ml)

        elif choice == "3":
            print("\n----- Daily Report -----")
            print(f"Daily Goal: {tracker.goal} ml")
            print(f"Total Drank: {tracker.get_total()} ml")
            print(f"Remaining: {tracker.get_remaining()} ml")
            print(f"Completion: {tracker.get_percentage():.2f}%")
            print("-------------------------\n")

        elif choice == "4":
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Try again.\n")

if __name__ == "__main__":
    main()
```
